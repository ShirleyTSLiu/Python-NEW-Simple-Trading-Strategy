### Load data

```python
data = pd.read_csv('csv.csv',header=0)
# pd.DataFrame(data)
```

```python
trade = []
for i in range(len(data)):
    row = []
    for j in range(0, 12):
        row.append(0)
    trade.append(row)
trade[0][0]='date'
trade[0][1]='close'
trade[0][2]='ma10'
trade[0][3]='buy'
trade[0][4]='hold'
trade[0][5]='sell'
trade[0][6]='number'
trade[0][7]='每日未實現損益'
trade[0][8]='累積未實現損益'
trade[0][9]='已實現損益'
trade[0][10]='value'
trade[0][11]='holding_days'
```

```python
# note that all ndarrays must be the same length!
inputs = {
    'open': data['open'],
    'high': data['high'],
    'low': data['low'],
    'close': data['close'],
}

# uses close prices
SMA10 = SMA(inputs, timeperiod=10, price='close')
```

```python
data['sma10'] = SMA10
```

```python
data = data.values.tolist()
```

```python
for i in range(1, len(data)):
    trade[i][0]=data[i][0]
    trade[i][1]=data[i][4]
    trade[i][2]=data[i][5]
```

#### Strategy: If Close price > SMA10 = Buy, else Sell

##### only long strategy

```python
for i in range(10, len(trade)):
    if (trade[i][1]>trade[i][2]) and (trade[i-1][4]==0):
        # close > ma10 and hold==0
        trade[i][3]=1
        trade[i][4]=1
        trade[i][5]=0
    elif (trade[i][1]<trade[i][2]) and (trade[i-1][4]==1):
        # close < ma10 and hold(t-1)==1
        trade[i][3]=0
        trade[i][4]=0
        trade[i][5]=1
    elif (trade[i-1][4]>0):
        # hold>0
        trade[i][3]=0
        trade[i][4]=1
        trade[i][5]=0
```

```python
pd.DataFrame(trade)
```

#### Holding days

```python
start = 0
end = 0
for i in range(1, len(trade)):
    if(trade[i][3]==1):
        trade[i][11] = 1
        start = datetime.datetime.strptime(trade[i][0], "%Y/%m/%d").date()
    elif((trade[i][4] > 0) or (trade[i][5] == 1))and(trade[i][3] == 0):
        end = datetime.datetime.strptime(trade[i][0], "%Y/%m/%d").date()
        delta = end - start
        trade[i][11] = delta.days+1
pd.DataFrame(trade)
```

```python
initial = 10000
```

```python
for i in range(10, len(trade)):
    if (trade[i][3]==1) and (trade[i][4]==1) and (trade[i][5]==0):
        # buy=1, hold=1, sell=0
        trade[i][6]=int(initial/trade[i][1]) #number
        trade[i][7]=0 #每日未實現損益
        trade[i][8]=0 #累積未實現損益
        trade[i][9]=0 #已實現損益
        trade[i][10]=trade[i][1]*trade[i][6] #value
    elif (trade[i][3]==0) and (trade[i][4]==1) and (trade[i][5]==0):
        # buy=0, hold=1, sell=0
        trade[i][6]=trade[i-1][6] #number
        trade[i][7]=int((trade[i][1]-trade[i-1][1])*trade[i-1][6]) #每日未實現損益
        trade[i][8]=trade[i-1][8]+trade[i][7] #累積未實現損益
        trade[i][9]=0 #已實現損益
        trade[i][10]=trade[i][1]*trade[i-1][6] #value
    elif (trade[i][3]==0) and (trade[i][4]==0) and (trade[i][5]==1):
        # buy=0, hold=0, sell=1
        trade[i][6]=0 #number
        trade[i][7]=int((trade[i][1]-trade[i-1][1])*trade[i-1][6]) #每日未實現損益
        trade[i][8]=trade[i-1][8]+trade[i][7] #累積未實現損益
        trade[i][9]=trade[i][8] #已實現損益
        trade[i][10]=trade[i][1]*trade[i-1][6] #value 
        initial=trade[i][10]
```
